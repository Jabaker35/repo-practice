```python
# merge, join and concatenate df
data1 = {'Name':['Alvin','Bobby'],
       'Age':[12,13],
       'Address':['Newark','Irvington'],
       'Qual':['7th','8th']}

data2 = {'Name':['Samuel','Robert'],
       'Age':[10,11],
       'Address':['Orange','Maplewood'],
       'Qual':['6th','9th']}

df = pd.DataFrame(data1,index=[0,1])
df1 = pd.DataFrame(data2,index=[2,3])

frame = [df,df1]
result = pd.concat(frame)
result
```

```python
# merge, join and concatenate df
data1 = {'Name':['Alvin','Bobby'],
       'Age':[12,13],
       'Address':['Newark','Irvington'],
       'Qual':['7th','8th']}

data2 = {'Name':['Samuel','Robert'],
       'Age':[10,11],
       'Address':['Orange','Maplewood'],
       'Qual':['6th','9th']}

df = pd.DataFrame(data1,index=[0,1])
df1 = pd.DataFrame(data2,index=[2,3])

result = df.append(df1)
result
```

```python
# merging df using unique key
data1 = {'key':['k0','k1'],
        'Name':['Alvin','Bobby'],
       'Age':[12,13]}
  
data2 = {'key':['k0','k1'],
        'Name':['Samuel','Robert'],
        'Age':[10,11]}

df = pd.DataFrame(data1)
df1 = pd.DataFrame(data2)

result = pd.merge(df,df1,on='key')
result
```
