```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
data = pd.read_csv('StudentsPerformance.csv')
data
```

```python
# 1. unique value in gender
# 2. unique value in race/ethnicity
# 3. remove race/ethnicity and parental level of education and assign to var student
# 4. calculate corr for student
# 5. pairplot on student
# 6. distplot on student and col = math score
```
