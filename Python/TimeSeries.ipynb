```python
# Additive Model: Y(t) = Trend(t) + Seasonality(t) + Cyclic Patterns(t) + (Residual(t): diff b/w expected and observed values)
# it is linear b/c changes over time are consistently made by same amount
# same frequency(width of cycle) and amplitude(height of cycle)
# sm.tsa.seasonal_decompose(tsa package has seasonal decompose function)
    # import seasonal_decompose
    # series = .....
    # result = seasonal_decompose(series,model='additive')
    # print(result.trend) (.seasonality, .residuals)
    # from matplotlib import pyplot
    # series = ....
    # result = seasonal_decompose(series,model='additive')
    # result.plot()

# Multiplicative model: same components ar multiplied together
    # if seasonality and residual are independent of trend: additive series
    # if seasonality and residual are dependent of trend: multiplicative series
    
# Additive trend is linear (straight line) and multilicative means there are changes to w's or h's of seasonal period over time
    # 1. magnitude of seasonal variations does not vary
    # 2. used where change is measured in absolute quantity

# Multiplicative Series:
    # 1. seasonal variations increase or decrease proportionally with increase or decrease in the level of series
    # 2. used where change is measured in perctange change (non-linear so it is a curved line)
    
# Choosing models:
    # choose multiplicative when magnitude of seasonal patterns in data depends upon the magnitude of data
    # choose additive when magnitude of seasonal patterns in data does not depends upon the magnitude of data
    # try both and pick the model with better results
```

```python
matplotlib.rcParams['figure.figsize'] = [12,8]
decomp = sm.tsa.seasonal_decompose(data,model='additive')
fig = decomp.plot()
```

```python
# Level: avg value in the series
# Trend: increasing or decreasing values in the series
# Seasonality: repeating short term cycle in the series
# Residual/Noise : random variation in the series
```

```python
# time series data comprised of linearly increasing trend from 1 to 99 with some random noise
# decompose it as an additive model
df = [i + randrange(10) for i in range(1,100)]
df
result = seasonal_decompose(df,model='additive',period=1)
result.plot()
```

```python
# multiplicative decomposition
df = [i**2 for i in range(1,100)]
result = seasonal_decompose(df,model='multiplicative',period=1)
result.plot()
pyplot.show()
```

```python
df = pd.read_csv('documents/dailysales.csv',parse_dates=['date'])
df
```

```python
df.dtypes
df.info()
df.isnull().sum()
df.head()
```

```python
df = df.groupby('date').sum()
df
```

```python
# aggregating data grouped by monthly sales
# A = means year end, MS = month start, M = month end, W = weekly, D = daily
# data reps daily sales aggregate it to monthly sales
df = df.resample(rule = 'M').sum()
df
```

```python
df.plot()
```

```python
# period = 4: quarterly frequency
seasonal_decompose(df,model='additive',period=4).plot()
```

```python
# splitting data
train = df[:21]
test = df[19:]
# shows the last 5 rows by defualt
train.tail()
```

```python
# Holt winter model is used when data has trend and seasonality but no randomness sometimes 
# (level,trend,seasonality)
# try all the possible combinations and use the model which gives better accuracy
hwmodel = ExponentialSmoothing(train.sales,trend='add',seasonal='mul',seasonal_periods=4).fit()
hwmodel
```

```python
# predict 5 months after july in training set
testpred = hwmodel.forecast(5)
testpred
```

```python
train['sales'].plot(legend=True,label='Train')
test['sales'].plot(legend=True,label='Test')
testpred.plot(legend=True,label='Predicted Test')
```

```python
# model evaluation
# lower the mean indicates predicted values are very 
np.sqrt(mean_squared_error(test,testpred))
```

```python
# final model
finalmodel = ExponentialSmoothing(df.sales,trend='add',seasonal='mul',seasonal_periods=4).fit()
finalmodel
```

```python
df = pd.read_csv('documents/GlobalLandTemperaturesByState.csv')
df
```

```python
df.head()
df.dtypes()
df.shape()
df.isnull().sum()
```

```python
# removing rows with null values
df = df.dropna()
df
```

```python
df.isnull().sum()
```

```python
# renaming columns 
# dt=Date, AverageTemperature=Avg_Temp,AverageTemperatureUncertainty=Avg_Uncert_Temp
df.rename(columns = {'dt':'Date','AverageTemperature':'Avg_Temp',
                        'AverageTemperatureUncertainty':'Avg_Uncert_Temp'},inplace=True)
df
```

```python
# convert Date into datetime and index as column
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date',inplace=True)
df
```

```python
df
```

```python
# create 'Year' col
df['Year'] = df.index.year
df
```

```python
latestdf = df.loc['1980':'2013']
latestdf.head()
```

```python
# find the average temp based on countries
latestdf[['Country','Avg_Temp']].groupby(['Country']).mean().sort_values('Avg_Temp')
```

```python
resampledf = latestdf[['Avg_Temp']].resample('A').mean()
resampledf.head()
```

```python
# plot resampledf
resampledf.plot(title='Temperature changes from 1980 to 2013',figsize=(8,5))
plt.ylabel('Temperature',fontsize=12)
plt.xlabel('Year',fontsize=12)
plt.legend
```

```python
decomp = seasonal_decompose(resampledf,period=3)
trend = decomp.trend
seasonal = decomp.seasonal
residual = decomp.resid
```

```python
plt.subplot(411)
plt.plot(resampledf)
plt.xlabel('Original')
plt.figure(figsize=(6,5))

plt.subplot(412)
plt.plot(trend)
plt.xlabel('Trend')
plt.figure(figsize=(6,5))

plt.subplot(413)
plt.plot(seasonal)
plt.xlabel('Seasonal')
plt.figure(figsize=(6,5))

plt.subplot(414)
plt.plot(residual)
plt.xlabel('Residual')
plt.figure(figsize=(6,5))

plt.tight_layout()
```
